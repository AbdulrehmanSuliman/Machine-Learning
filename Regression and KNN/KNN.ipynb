### Question 2
What is the class distribution? (i.e. how many instances of `malignant` and how many `benign`?)

### Question 3
Split the DataFrame into `X` (the data) and `y` (the labels).

```python
def answer_three():
    cancerdf = answer_one()
    
    # Your code here
    X= cancerdf.loc[:,cancerdf.columns != 'target']
    y= cancerdf['target']
    return X, y
answer_three()
```

### Question 4
Using `train_test_split`, split `X` and `y` into training and test sets

```python
from sklearn.model_selection import train_test_split

def answer_four():
    X, y = answer_three()
    
    # Your code here
    X_train,X_test,y_train,y_test=train_test_split(X, y, random_state=0)
    
    return X_train, X_test, y_train, y_test
answer_four()
```

### Question 9
##### Accuracy Plot
* Create `mal_train_X`, `mal_train_y`,`ben_train_X`,`ben_train_y`, `mal_test_X`,`mal_test_y`, `ben_test_X` and `ben_test_y` portions of the dataset. Where `mal_train_X` contains all training samples that belongs to malignant class while `mal_train_y` contains all labels of the samples, and so on for other variables.

* Calculate Accuracy scores for each case and save all the results in a list called `scores` using knn model from question 5.

Try using the plotting function below to visualize the differet predicition scores between training and test sets, as well as malignant and benign cells.

```python
def accuracy_plot():
    import matplotlib.pyplot as plt

    %matplotlib notebook

    X_train, X_test, y_train, y_test = answer_four()

    # Find the training and testing accuracies by target value (i.e. malignant, benign)
    
    # #################################################################################
    # ###################### Write your code here #####################################
    # #################################################################################
    mal_train_index = y_train==0
    mal_test_index = y_test==0
    ben_train_index = y_train==1
    ben_test_index = y_test==1
    mal_train_X = X_train[mal_train_index]
    mal_train_y = y_train[mal_train_index]
    ben_train_X = X_train[ben_train_index]
    ben_train_y = y_train[ben_train_index]
    mal_test_X = X_test[mal_test_index]
    mal_test_y = y_test[mal_test_index]
    ben_test_X = X_test[ben_test_index]
    ben_test_y = y_test[ben_test_index]

    knn = answer_five()
    mal_training_score=knn.score(mal_train_X, mal_train_y)
    ben_training_score=knn.score(ben_train_X, ben_train_y)
    mal_testing_score=knn.score(mal_test_X, mal_test_y)
    ben_testing_score=knn.score(ben_test_X, ben_test_y)

    scores = [mal_training_score,ben_training_score,mal_testing_score, ben_testing_score]


    plt.figure()

    # Plot the scores as a bar chart
    bars = plt.bar(np.arange(4), scores, color=['#4c72b0','#4c72b0','#55a868','#55a868'])

    # directly label the score onto the bars
    for bar in bars:
        height = bar.get_height()
        plt.gca().text(bar.get_x() + bar.get_width()/2, height*.90, '{0:.{1}f}'.format(height, 2), 
                     ha='center', color='w', fontsize=11)

    # remove all the ticks (both axes), and tick labels on the Y axis
    plt.tick_params(top='off', bottom='off', left='off', right='off', labelleft='off', labelbottom='on')

    # remove the frame of the chart
    for spine in plt.gca().spines.values():
        spine.set_visible(False)

    plt.xticks([0,1,2,3], ['Malignant\nTraining', 'Benign\nTraining', 'Malignant\nTest', 'Benign\nTest'], alpha=0.8);
    plt.title('Training and Test Accuracies for Malignant and Benign Cells', alpha=0.8)
```

```python
accuracy_plot() 
```

```python
X=irisDF.loc[:,irisDF.columns != 'target']
y=irisDF['target']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X, y,test_size=0.3 ,random_state=0)
X_train,X_test,y_train,y_test
```

```python
from matplotlib import cm

cmap = cm.get_cmap('gnuplot')
scatter = pd.plotting.scatter_matrix(X_train, c= y_train, marker = 'o', s=40, hist_kwds={'bins':15}, figsize=(9,9), cmap=cmap)
```

```python
max_score=0
feature_1=''
feature_2=''
knn=KNeighborsClassifier(n_neighbors=5)
for i in range(len(iris.feature_names)-1):
    for j in range(i+1,len(iris.feature_names)): 
        train_features=X_train[[iris.feature_names[i],iris.feature_names[j]]]
        two_features_train=train_features.copy()
        test_features=X_test[[iris.feature_names[i],iris.feature_names[j]]]
        two_features_test=test_features.copy()
        
        knn.fit(two_features_train,y_train)
        knn.predict(two_features_test)
        score=knn.score(two_features_test,y_test)
        
        if score >= max_score:
            feature_1=iris.feature_names[i]
            feature_2=iris.feature_names[j]
            max_score=score
feature_1,feature_2,max_score
```
