```python
class Node ():
    def __init__(self,attribute=None,threshold=None,left=None,right=None, gain=None ,descision=None):
        self.attribute = attribute
        self.threshold = threshold
        self.left = left
        self.right = right
        self.gain = gain
        self.descision = descision
        
```

```python
class DescisionTree():
    def __init__(self,maxDepth=2):
        self.maxDepth= maxDepth
        self.root= None
    
    
    def getPartition(self, data,features):
        partition=[]
        maxGain=np.NINF
        for feature in range(features):
            thresholds=np.unique(data[:,feature])
            for threshold in thresholds:
                left_data=np.array([samplefeature for samplefeature in data if samplefeature[feature] <= threshold])
                right_data=np.array([samplefeature for samplefeature in data if samplefeature[feature] > threshold])
                if len(left_data) > 0 and len(right_data) > 0:

                    gain = self.calculateGain(data[:,-1],left_data[:,-1],right_data[:,-1])
                    if gain > maxGain:
                        partition = [feature, threshold, left_data, right_data, gain ]
                        maxGain = gain
        return partition
    
    def calculateGain(self,root,left,right):
        pLeft = len(left)/len(root)
        pRight = len(right)/len(root)
        return self.entropy(root)-(pLeft*self.entropy(left)+pRight*self.entropy(right)) 
    
    def entropy(self,dataLabels):
        entropy=[]
        for classType in np.unique(dataLabels):
            probability=len(dataLabels[dataLabels==classType])/len(dataLabels)
            entropy.append(-probability*np.log2(probability))
        return sum(entropy)
    
    def buildTree(self,data, depth=0):
        X=data[:,:-1]
        y=data[:,-1]
        features =X.shape[1]
        if depth <= self.maxDepth:
            partition = self.getPartition(data,features)
            if partition[4] > 0:
                left = self.buildTree(partition[2],depth=depth+1)
                right= self.buildTree(partition[3],depth=depth+1)
                return Node(partition[0],partition[1],left,right,partition[4])
        unique,counts = np.unique(y,return_counts=True)
        return Node(descision=unique[np.argmax(counts)])
    
    def fit(self, X,y):
        dataset=np.concatenate((X,y),axis=1)
        self.root=self.buildTree(dataset)

    def getDecision(self,test, tree):
            if tree.descision ==None:
                if test[tree.attribute] <=tree.threshold:
                    return self.getDecision(test, tree.left)
                else:
                    return self.getDecision(test,tree.right)
            else:
                return tree.descision

    def predict(self,test):
          
        preditions = [self.getDecision(x, self.root) for x in test]
        return preditions
```

```python
from sklearn.model_selection import train_test_split
classifier= DescisionTree()
X = df.iloc[:, :-1].values
Y = df.iloc[:, -1].values.reshape(-1,1)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=0)
classifier.fit(X_train,Y_train)
```

```python
Y_pred = classifier.predict(X_test) 
from sklearn.metrics import accuracy_score
accuracy_score(Y_test, Y_pred)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state=0)
clf.fit(X_train,Y_train)
prd=clf.predict(X_test)
accuracy_score(Y_test, prd)
```
